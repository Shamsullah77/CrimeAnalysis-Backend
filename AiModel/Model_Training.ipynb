```python
# Define features and target variables
features = traindf[['date', 'hour_of_day', 'location',
                    'victim_gender', 'victim_age', 'perpetrator_gender', 'perpetrator_age',
                    'weapon', 'time_of_day', 'day_of_week', 'month', 'season',
                    'victim_age_group', 'perpetrator_age_group', 'weather_condition',
                    'economic_index', 'education_index']]

target = traindf[['crime_type', 'location']]

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Output the results
print("X_train:")
print(x_train.head())

print("\nX_test:")
print(x_test.head())

print("\nY_train:")
print(y_train.head())

print("\nY_test:")
print(y_test.head())
```

Column Transformer Setup

```python
# Identify numerical and categorical columns
numerical_cols = x_train.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = x_train.select_dtypes(include=['category', 'object']).columns

print("The numerical columns:", numerical_cols)
print("The categorical columns:", categorical_cols)

# Create a column transformer for preprocessing
col_tsf = ColumnTransformer(
    transformers=[
        ('num', Pipeline([('imputer', SimpleImputer(strategy='median')), ('scaler', StandardScaler())]), numerical_cols),
        ('cat', Pipeline([('imputer', SimpleImputer(strategy='most_frequent')), ('onehot', OneHotEncoder(handle_unknown='ignore'))]), categorical_cols)
    ],
    remainder='passthrough'
)

# Fit and transform the training data
x_preprocessed_train = col_tsf.fit_transform(x_train)

# Transform the test data
x_preprocessed_test = col_tsf.transform(x_test)

# Save the column transformer
joblib.dump(col_tsf, 'column_transformer.pkl')

print("Preprocessing complete.")
```

Hyperparameter Tuning and Model Training

```python
# Define the RandomForestClassifier
RFC = RandomForestClassifier()

# Define the parameter grid with optimized values
prams = {
    "max_features": uniform(0.1, 0.5),
    "n_estimators": randint(50, 200),
    "max_depth": randint(10, 30),
    "min_samples_split": randint(2, 10)
}

# Train and evaluate separate classifiers for crime type and location
def train_and_evaluate(x_train, y_train, prams, model_name):
    rnd_search = RandomizedSearchCV(RFC, prams, n_iter=30, cv=3, n_jobs=-1, random_state=42)
    rndx = rnd_search.fit(x_train, y_train)
    
    # Save the trained model
    joblib.dump(rndx, f'trained_random_forest_model_for_{model_name}.pkl')

    return rndx

# Separate the target variables into crime type and location
y_train_type = y_train['crime_type']
y_train_location = y_train['location']

# Train and save models
rndx_type = train_and_evaluate(x_preprocessed_train, y_train_type, prams, "crime_type")
rndx_location = train_and_evaluate(x_preprocessed_train, y_train_location, prams, "location")

print("Hyperparameter tuning complete.")
```

```python
numeric_features = ['victim_age', 'perpetrator_age', 'economic_index', 'education_index']
categorical_features = ['location', 'victim_gender', 'perpetrator_gender', 'weapon', 'weather_condition']

class DataPreprocessor(BaseEstimator, TransformerMixin):
    def __init__(self, numeric_features, categorical_features):
        self.numeric_features = numeric_features
        self.categorical_features = categorical_features
        
        self.numeric_pipeline = Pipeline([
            ('imputer', SimpleImputer(strategy='mean')),
            ('scaler', StandardScaler())
        ])
        
        self.categorical_pipeline = Pipeline([
            ('imputer', SimpleImputer(strategy='most_frequent')),
            ('onehot', OneHotEncoder(handle_unknown='ignore'))
        ])
        
        self.preprocessor = ColumnTransformer([
            ('numeric', self.numeric_pipeline, self.numeric_features),
            ('categorical', self.categorical_pipeline, self.categorical_features)
        ])
        
    def fit(self, X, y=None):
        self.preprocessor.fit(X)
        return self
    
    def transform(self, X):
        return self.preprocessor.transform(X)

data_preprocessor = DataPreprocessor(numeric_features, categorical_features)
X_processed = data_preprocessor.fit_transform(traindf)
```

```python
X = traindf.drop(['crime_type', 'location'], axis=1)
y_crime_type = traindf['crime_type']
y_location = traindf['location']

print("Null values in Crime Type:")
print(y_crime_type.isnull().any())

print("\nNull values in Location:")
print(y_location.isnull().any())
```

```python
features = traindf[['date', 'hour_of_day', 'location',
                    'victim_gender', 'victim_age', 'perpetrator_gender', 'perpetrator_age',
                    'weapon', 'time_of_day', 'day_of_week', 'month', 'season',
                    'victim_age_group', 'perpetrator_age_group', 'weather_condition',
                    'economic_index', 'education_index']]
target = traindf[['crime_type', 'location']]

x_train, x_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

print("X_train:")
print(x_train.head())

print("\nX_test:")
print(x_test.head())

print("\nY_train:")
print(y_train.head())

print("\nY_test:")
print(y_test.head())
```

```python
numerical_cols = x_train.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = x_train.select_dtypes(include=['category', 'object', 'bool']).columns

print("The numerical columns:", numerical_cols)
print("The categorical columns:", categorical_cols)

col_tsf = ColumnTransformer([
    ('num', Pipeline([
        ('imputer', SimpleImputer(strategy='median')),
        ('scaler', StandardScaler())
    ]), numerical_cols),
    ('cat', Pipeline([
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('onehot', OneHotEncoder(handle_unknown='ignore'))
    ]), categorical_cols)
], remainder='passthrough')

x_preprocessed_train = col_tsf.fit_transform(x_train)
x_preprocessed_test = col_tsf.transform(x_test)

joblib.dump(col_tsf, 'column_transformer.pkl')
print("Preprocessing complete.")

```

```python
prams = {
    "max_features": uniform(0.1, 0.5),
    "n_estimators": randint(50, 200),
    "max_depth": randint(10, 30),
    "min_samples_split": randint(2, 10)
}

def train_and_evaluate(x_train, y_train, prams, model_name):
    rnd_search = RandomizedSearchCV(RandomForestClassifier(), prams, n_iter=30, cv=3, n_jobs=-1, random_state=42)
    rndx = rnd_search.fit(x_train, y_train)
    joblib.dump(rndx, f'trained_random_forest_model_for_{model_name}.pkl')
    return rndx

y_train_type = y_train['crime_type']
y_train_location = y_train['location']

rndx_type = train_and_evaluate(x_preprocessed_train, y_train_type, prams, "crime_type")
rndx_location = train_and_evaluate(x_preprocessed_train, y_train_location, prams, "location")

print("Hyperparameter tuning complete.")
```
